```python
def simple_ellipse(x0, y0, a, b, phi=0, resolution=1000):
    """Draw a simple ellipse with shifted center."""
    t = np.linspace(0, 2 * np.pi, resolution)
    x = x0 + a * np.cos(t) * np.cos(phi) - b * np.sin(t) * np.sin(phi)
    y = y0 + a * np.cos(t) * np.sin(phi) + b * np.sin(t) * np.cos(phi)
    
    return x, y
```

```python
x, y = simple_ellipse(2, 5, 10, 5, phi=45.0)

plt.plot(x, y)
```
