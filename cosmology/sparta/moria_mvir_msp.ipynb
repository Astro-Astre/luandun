```python
cat = '/Users/song/Downloads/moria_0.7603.hdf5'

field_list = ['id', 'parent_id_cat', 'parent_id_Rsp-apr-p50', 'Mvir_all_spa', 'Mvir_bnd_cat', 'Mvir_tcr_spa',
              'Msp-apr-p50', 'Msp-apr-p90', 'Acc_Rate_2*Tdyn', 'scale_of_last_MM']

data = moria.load(cat, field_list=field_list, log_level=1)
```

```python
mask = (data['parent_id_cat'] == -1) & (data['Mvir_bnd_cat'] >= 1e13)
print(mask.sum())
```
